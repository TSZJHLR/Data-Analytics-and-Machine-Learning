## Most Popular Treks and their Average Cost

```python
df["trek"].value_counts()
```

```python
df["trek_peaks"].value_counts()
```

The code analyzes trekking data to identify the top 5 most frequent peaks. It then calculates the average cost for each of these top peaks and visualizes this information in a bar chart, providing insights into the most popular and potentially most expensive trekking destinations.

```python
top_peaks = df["trek_peaks"].value_counts().nlargest(5).index
top_peaks_df = df[df["trek_peaks"].isin(top_peaks)]
display(top_peaks_df)
```

```python
avg_cost_per_peak = top_peaks_df.groupby("trek_peaks")["cost_in_npr"].mean().reset_index()
fig_cost = px.bar(avg_cost_per_peak,
                  x='trek_peaks',
                  y='cost_in_npr',
                  title="Average Cost for Top Peaks",
                  labels={'trek_peaks': 'Trek Peaks', 'cost_in_npr': 'Average Cost in NPR'})
fig_cost.update_layout({"bargap": 0.2})
fig_cost.show()
```

```python
df.columns
```

The code analyzes trekking data to determine the average time required for the top 5 most popular peaks. It then visualizes this information in a bar chart, providing insights into the duration of treks for each peak.

```python
avg_time_per_peak = top_peaks_df.groupby("trek_peaks")["time_in_days"].mean().reset_index()
fig_time = px.bar(avg_time_per_peak,
                  x='trek_peaks',
                  y='time_in_days',
                  title="Average Time for Top Peaks",
                  labels={'trek_peaks': 'Trek Peaks', 'time_in_days': 'Average Time in Days'})
fig_time.update_layout({"bargap": 0.2})
fig_time.show()
```

### Cost Heatmap Based on Altitude and Month

This code snippet generates a density heatmap using Plotly Express to visualize the average cost of treks (in NPR) across different altitude ranges and travel months. It first categorizes the `max_altitude` values into specified ranges using `pd.cut`. The data is then grouped by altitude range and traveled month to calculate the mean cost. The heatmap displays travel months on the x-axis and altitude ranges on the y-axis, with color intensity representing the average cost. The plot is titled "Average Cost by Altitude Range and Month," effectively highlighting cost trends in relation to altitude and season.

```python
df['altitude_range'] = pd.cut(df['max_altitude'], bins=[0, 2000, 3000, 4000, 5000, 6000, 7000],
                              labels=['<2000m', '2000-3000m', '3000-4000m', '4000-5000m', '5000-6000m', '6000m+'])

cost_heatmap = df.groupby(['altitude_range', 'traveled_month'])['cost_in_npr'].mean().reset_index()
fig = px.density_heatmap(cost_heatmap,
                         x='traveled_month',
                         y='altitude_range',
                         z='cost_in_npr',
                         title="Average Cost by Altitude Range and Month",
                         labels={'traveled_month': 'Travel Month', 'altitude_range': 'Altitude Range', 'cost_in_npr': 'Avg Cost in NPR'},)
fig.show()
```

## Income Analysis

This code snippet creates a multi-panel figure using Plotly to analyze trekking destination trips based on annual income. First, it categorizes annual incomes into defined groups using `pd.cut` and calculates the trip counts for each income group and trekking peak.

The figure consists of three subplots:

1. **Income Distribution by Trekking Destination**: A stacked bar chart showing the distribution of trips to various trekking destinations across different income groups.
2. **Average Trip Count by Income Group**: A bar chart displaying the average number of trips for each income group.
3. **Income vs. Trip Count**: A scatter plot illustrating the relationship between annual income and the number of trips to trekking destinations, with each point representing a specific trekking peak.

The overall title of the figure is "Income Analysis for Trekking Destinations," and the layout is configured for clarity with appropriate axis labels. This comprehensive visualization effectively highlights trends and patterns in trekking behaviors related to income levels.

```python
sorted(df["annual_income"].unique())
```

```python
income_bins = [0, 500000, 1000000, 1500000, 2000000]
income_labels = ['<500k', '500k-1M', '1M-1.5M', '1.5M+']
df['income_group'] = pd.cut(df['annual_income'], bins=income_bins, labels=income_labels, right=False)

income_distribution = df.groupby(['income_group', 'trek_peaks']).size().reset_index(name='trip_count')

average_trip_count = df.groupby('income_group')['trek_peaks'].count().reset_index(name='average_trip_count')

fig = make_subplots(
    rows=3, cols=1,
    subplot_titles=("Income Distribution by Trekking Destination",
                    "Average Trip Count by Income Group",
                    "Income vs. Trip Count"),
    vertical_spacing=0.15
)

for income_group in income_distribution['income_group'].unique():
    income_data = income_distribution[income_distribution['income_group'] == income_group]
    fig.add_trace(
        go.Bar(
            x=income_data['trek_peaks'],
            y=income_data['trip_count'],
            name=income_group,
            hoverinfo='x+y'
        ),
        row=1, col=1
    )

fig.add_trace(
    go.Bar(
        x=average_trip_count['income_group'],
        y=average_trip_count['average_trip_count'],
        name='Average Trip Count',
        hoverinfo='x+y'
    ),
    row=2, col=1
)

# Scatter Plot: Income vs. Trip Count
scatter_data = df.groupby(['annual_income', 'trek_peaks']).size().reset_index(name='trip_count')
fig.add_trace(
    go.Scatter(
        x=scatter_data['annual_income'],
        y=scatter_data['trip_count'],
        mode='markers',
        marker=dict(size=5, opacity=1, color='blue'),
        text=scatter_data['trek_peaks'],
        hoverinfo='text+y'
    ),
    row=3, col=1
)

fig.update_layout(
    title="Income Analysis for Trekking Destinations",
    xaxis_title="Trekking Destination",
    yaxis_title="Trip Count / Average Trip Count",
    height=1200,
    showlegend=True
)

# Show the plot
fig.show()
```
