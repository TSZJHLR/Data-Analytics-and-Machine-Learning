### removing black spaces, renaming cloumns, lowercase

```python
def format_column(col):
    col = re.sub(r'\s+', '', col)
    col = re.sub('(?<!^)(?=[A-Z])', '_', col).lower()
    return col

df3.columns = [format_column(col) for col in df3.columns]
print(df3.columns)
```

### renaming to single words

```python
def refine_accomodation(value):
  if isinstance(value, str):
    value = value.replace("Teahouse", "Tea")
    value = value.replace("Guesthouse", "Guest")
    value = value.replace("Luxury Lodge", "Lodge")
    value = value.replace("Guest House", "Guest")
  return value

df3['accomodation_1'] = df3['accomodation_1'].apply(refine_accomodation)
df3['accomodation_2'] = df3['accomodation_2'].apply(refine_accomodation)
```

### best time to travel

```python
df4 = df3.copy(deep=True)
```

```python
print(df4['best_travel_time'].unique())
```

```python
def clean_travel_time(value):
  value = value.replace(".", "")
  value = value.replace("March", "Mar")
  value = value.replace("April", "Apr")
  value = value.replace("Setpt", "Sep")
  value = value.replace("Sept", "Sep")
  value = value.replace(" ", "")
  value = re.sub(r'(\w+)-(\w+)', r'\1 - \2', value)
  value = re.sub(r'&(\w+)', r' & \1', value)
  return value

df4['best_travel_time'] = df4['best_travel_time'].apply(clean_travel_time)

print(df4['best_travel_time'].unique())
```

```python
df4[['best_travel_time_1','best_travel_time_2']] = df4['best_travel_time'].str.split(' & ',expand=True)
df4['best_travel_time_1'] = df4['best_travel_time_1'].str.strip()
df4['best_travel_time_2'] = df4['best_travel_time_2'].str.strip()
df4 = df4.drop('best_travel_time', axis=1)
```

```python
print(df4['best_travel_time_1'].unique())
print(df4['best_travel_time_2'].unique())
```

```python
def season_to_month(season):
  season_month = {
      'Mar - May': 'Spring',
      'Jun - Aug': 'Summer',
      'Sep - Nov': 'Autumn',
      'Dec - Feb': 'Winter',
      'Sep - Dec': 'Autumn-Winter',
      'Jan - May': 'Spring-Winter',
      'Apr - May': 'Spring-Summer',
      'Mar - Nov': 'Non-Winter'
  }
  return season_month.get(season)

df4['best_travel_time_1'] = df4['best_travel_time_1'].apply(season_to_month)
df4['best_travel_time_2'] = df4['best_travel_time_2'].apply(season_to_month)
```

```python
# @title time_in_days vs cost_in_usd

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['time_in_days']
  ys = series['cost_in_usd']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df4.sort_values('time_in_days', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('family_members')):
  _plot_series(series, series_name, i)
  fig.legend(title='family_members', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('time_in_days')
_ = plt.ylabel('cost_in_usd')
```
